```python
user_movies_df = movies_df.pivot(index = 'userId',
                                columns = 'movie',
                                values = 'rating')
```

```python
#Impute those NAN's with 0 values
user_movies_df.fillna(0, inplace = True)
```

```python
user_movies_df
```

```python
#Calculating cosine similarity
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1-pairwise_distances( user_movies_df.values,metric = 'cosine')
```

```python
user_sim
```

```python
#store the result in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
def get_reco(customer_id):
    '''
    let's assume customer_id id 8
    input is customer_id
    input is customer_id 8
    tem is going to contain all the other customers who are similar to customer 8
    movie list --> movies watched by top 100 similar customers to 8
    all set(movie_list)--> all the movies which are watched by 8'''
    
    tem= list(user_sim_df.sort_values([customer_id],ascending = False).head(100).index)
    #print(tom)
    movie_list = []
    for i in tem:
        movie_list = movie_list+list(movies_df[movies_df["userId"]==i]["movie"])
    #print(movie_list)
    return set(movie_list)-set(movies_df[movies_df["userId"]==customer_id]["movie"])
```
